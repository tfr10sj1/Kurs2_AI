```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris = pd.read_csv("iris_csv.csv")
```

```python
iris
```

```python
plt.scatter(iris["sepallength"], iris["sepalwidth"], color="red")
plt.scatter(iris["petallength"], iris["petalwidth"], color="blue")
plt.title("Petals and Sepals")
plt.xlabel("Length")
plt.ylabel("Width")
plt.show()
```

```python
setosa = iris.loc[iris['class'] == "Iris-setosa"]
virginica = iris.loc[iris['class'] == "Iris-virginica"]
versicolor = iris.loc[iris['class'] == "Iris-versicolor"]
```

```python
plt.scatter(setosa["sepallength"], setosa["sepalwidth"], color="red")
plt.scatter(virginica["sepallength"], virginica["sepalwidth"], color="blue")
plt.scatter(versicolor["sepallength"], versicolor["sepalwidth"], color="green")
plt.title("Sepals by class")
plt.xlabel("Length")
plt.ylabel("Width")
plt.show()
```
